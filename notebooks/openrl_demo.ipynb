## Training

Users can train CartPole agent with `openrl` with a few lines:

```python
import numpy as np
from openrl.envs.common import make
from openrl.modules.common import PPONet as Net
from openrl.runners.common import PPOAgent as Agent
# create environment
env = make("CartPole-v1", env_num=9)
# create network
net = Net(env)
# create training agent
agent = Agent(net)
# begin to train!
agent.train(total_time_steps=20000)
env.close()
```

## Test

After the training, users can test the agent with:

```python
 # Begin to test
env = make("CartPole-v1", render_mode="group_human", env_num=9, asynchronous=True)
agent.set_env(env)
obs, info = env.reset()
done = False
step = 0
while not np.any(done):
    # The agent predicts the next action based on environmental observations.
    action, _ = agent.act(obs, deterministic=True)
    obs, r, done, info = env.step(action)
    step += 1
    print(f"{step}: reward:{np.mean(r)}")
env.close()
```
